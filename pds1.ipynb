**1. What are sequence data types in Python?
   Name them and explain how indexing and slicing work across different sequence types?**

   ANS :
   Sequence data types in Python are ordered collections of items, where each item can be accessed by its position (index). They are fundamental to Python programming and are used to store multiple values in a structured way.

### Common Sequence Data Types:

1.  **Strings (`str`)**: An immutable sequence of Unicode characters.
    *   Example: `"hello"`

2.  **Lists (`list`)**: A mutable (changeable) sequence of items, which can be of different data types.
    *   Example: `[1, "apple", 3.14]`

3.  **Tuples (`tuple`)**: An immutable (unchangeable) sequence of items, similar to lists but cannot be modified after creation.
    *   Example: `(1, "banana", 5.0)`

4.  **Ranges (`range`)**: An immutable sequence of numbers, commonly used for looping a specific number of times. It's a special type of sequence.
    *   Example: `range(0, 5)`

### Indexing:

Indexing is the process of accessing individual elements within a sequence using their position. Python uses zero-based indexing, meaning the first element is at index `0`, the second at `1`, and so on.

*   **Positive Indexing**: Starts from the beginning of the sequence (`0`, `1`, `2`, ...).
*   **Negative Indexing**: Starts from the end of the sequence (`-1` for the last element, `-2` for the second to last, etc.).

**How it works:**

*   **Syntax**: `sequence_name[index]`

**Example (using a list):**

```python
my_list = ["apple", "banana", "cherry", "date"]
print(my_list[0])    # Output: apple (first element)
print(my_list[2])    # Output: cherry (third element)
print(my_list[-1])   # Output: date (last element)
print(my_list[-3])   # Output: banana (third from last)
```

### Slicing:

Slicing allows you to extract a portion (a "slice") of a sequence. It creates a new sequence containing elements from a specified start index up to (but not including) an end index.

*   **Syntax**: `sequence_name[start:end:step]`
    *   `start`: The starting index of the slice (inclusive). If omitted, it defaults to `0`.
    *   `end`: The ending index of the slice (exclusive). If omitted, it defaults to the end of the sequence.
    *   `step`: The step or stride of the slice. If omitted, it defaults to `1`.

**How it works:**

**Example (using a list):**

```python
my_list = ["apple", "banana", "cherry", "date", "elderberry", "fig"]

print(my_list[1:4])     # Output: ['banana', 'cherry', 'date'] (elements from index 1 to 3)
print(my_list[:3])      # Output: ['apple', 'banana', 'cherry'] (elements from start to index 2)
print(my_list[2:])      # Output: ['cherry', 'date', 'elderberry', 'fig'] (elements from index 2 to end)
print(my_list[::2])     # Output: ['apple', 'cherry', 'elderberry'] (every second element)
print(my_list[::-1])    # Output: ['fig', 'elderberry', 'date', 'cherry', 'banana', 'apple'] (reversed sequence)
print(my_list[1:5:2])   # Output: ['banana', 'date'] (elements from index 1 to 4, with a step of 2)
```

### Indexing and Slicing Across Different Sequence Types:

Indexing and slicing work consistently across all sequence data types (strings, lists, and tuples) in Python. The key difference lies in mutability:

*   **Strings and Tuples (Immutable)**:
    *   You can index and slice them to retrieve parts of the sequence.
    *   You **cannot** use indexing or slicing to change individual elements or parts of the sequence directly.
    *   Example: `my_string[0] = 'H'` or `my_tuple[1] = 'grape'` would raise a `TypeError`.

*   **Lists (Mutable)**:
    *   You can index and slice them to retrieve parts of the sequence.
    *   You **can** use indexing or slicing to modify existing elements or replace entire sections of the list.
    *   Example: `my_list[0] = "orange"` would change the first element.
    *   Example: `my_list[1:3] = ["blueberry", "mango"]` would replace elements at index 1 and 2.

**Example (Strings):**

```python
my_string = "Python"
print(my_string[0])     # Output: P
print(my_string[1:4])   # Output: yth
```

**Example (Tuples):**

```python
my_tuple = (10, 20, 30, 40)
print(my_tuple[1])      # Output: 20
print(my_tuple[2:])     # Output: (30, 40)
```

This consistent behavior makes it easy to work with different types of sequences once you understand the basic principles of indexing and slicing.

Ans : Differences Between Lists and Tuples:

| Feature           | List (`list`)                                     | Tuple (`tuple`)                                       |
| :---------------- | :------------------------------------------------ | :---------------------------------------------------- |
| **Mutability**    | Mutable (can be changed after creation)           | Immutable (cannot be changed after creation)          |
| **Syntax**        | Defined using square brackets `[]`                | Defined using parentheses `()`                        |

### Why Prefer a Tuple Over a List in Real-World Applications?

Despite lists being more flexible due to their mutability, tuples are often preferred in specific real-world scenarios for several reasons:

1.  **Data Integrity and Immutability**: When you want to ensure that the data stored in a collection remains constant and cannot be accidentally modified, tuples are the perfect choice. This is crucial for:
    *   **Configuration settings**: Storing application settings that shouldn't change during runtime.
    *   **Database records**: Representing a fixed row of data from a database query, where each field should retain its original value.
    *   **Coordinates/Points**: Storing (x, y) coordinates or RGB color values, which are inherently fixed.

2.  **Performance Optimization**: Because tuples are immutable, Python can make certain optimizations. This leads to:
    *   **Faster Processing**: Tuples are generally faster to iterate over and access elements from compared to lists, which can be beneficial in performance-critical applications or when dealing with large datasets.
    *   **Memory Efficiency**: Tuples consume less memory than lists because they don't need to allocate extra space for future modifications.

3.  **Use as Dictionary Keys**: Due to their immutability, tuples can be used as keys in dictionaries, whereas lists cannot. This is useful when you need composite keys (keys made of multiple values)

### Differences Between Python Lists and the `array` Module Arrays:

While both lists and arrays (from the `array` module) are sequence data types in Python used to store collections of items, they have fundamental differences that dictate their appropriate use cases.

| Feature           | List (`list`)                                     | Array (`array.array`)                                 |
| :---------------- | :------------------------------------------------ | :---------------------------------------------------- |
| **Data Type**     | Can store heterogeneous data types (e.g., int, string, float in the same list). | Stores homogeneous data types (all elements must be of the same specified type). |
| **Flexibility**   | Highly flexible; easy to modify, add, or remove elements. | Less flexible due to fixed data type; operations are type-specific. |
| **Memory Usage**  | Consumes more memory as each element is a Python object, storing both value and type information. | Consumes less memory as elements are stored compactly as raw C-style data types. |
| **Performance**   | Slower for numerical operations due to type checking and object overhead. | Faster for numerical operations on large datasets due to contiguous memory allocation and direct mapping to C types. |
| **Operations**    | Rich set of built-in methods (append, insert, remove, sort, etc.) applicable to any object type. | Fewer methods, primarily focused on basic array operations and conversions. |
| **Use Cases**     | General-purpose collection for mixed data.        | Efficient storage and manipulation of large sequences of numbers (e.g., sensor data, image pixel data). |

### When Should You Use the `array` Module Instead of a List?

YouYou should consider using the `array` module primarily when you need to store a **large number of homogeneous basic numeric data types** and are concerned about **memory efficiency** and **performance for numerical operations**.

Here are some specific scenarios:

1.  **Memory Efficiency**: If you have millions of integers or floats, `array.array` will consume significantly less memory than a Python list because it stores them as compact C-style data types, rather than Python objects.

2.  **Performance for Numerical Operations**: When performing mathematical operations (like addition, subtraction, multiplication) on a large sequence of numbers, the `array` module can be faster because the data is stored contiguously in memory, allowing for more efficient processing by the CPU.

3.  **Interfacing with C/C++ Libraries**: When working with external C or C++ libraries that expect C-style arrays as input, `array.array` provides a convenient and efficient way to transfer data without excessive copying or conversion overhead.

4.  **Disk Storage**: For storing sequences of numbers directly to disk and reading them back efficiently (e.g., binary file I/O), `array.array` is often preferred over lists.

**Example of `array.array`:**

```python
import array

# 'i' specifies signed integer type
```

**4. Explain how dictionaries work internally in Python.
   Why are dictionary lookups faster compared to lists or tuples?**
   
ANS :

### How Dictionaries Work Internally (Hash Tables):

Python dictionaries are implemented as **hash tables**. A hash table is a data structure that maps keys to values using a special function called a **hash function**. Here's a simplified breakdown:

1.  **Hash Function**: When you add a key-value pair to a dictionary, Python first takes the key and passes it through a **hash function**. This function converts the key into a fixed-size integer (the hash value).
    *   **Requirement**: Keys must be *hashable*, meaning their hash value doesn't change during their lifetime, and they can be compared to other objects. Immutable types like numbers, strings, and tuples are hashable. Mutable types like lists and other dictionaries are not.

2.  **Index Calculation**: The hash value is then used to calculate an index (a bucket or slot) in an internal array where the key-value pair will be stored. This allows for direct access to the potential location of the data.

3.  **Collision Handling**: It's possible for two different keys to produce the same hash value, or for two different hash values to map to the same index (a "collision"). Python uses various techniques to handle collisions, often by storing multiple key-value pairs that map to the same bucket in a linked list-like structure within that bucket, or by probing for the next available slot.

4.  **Storing Data**: The key-value pair is then stored in the calculated bucket. To ensure uniqueness and handle collisions, Python stores both the original key and its value.

5.  **Lookup Process**: When you try to retrieve a value using a key, Python repeats the process:
    *   It calculates the hash value of the provided key.
    *   It uses this hash value to jump directly to the corresponding bucket in the internal array.
    *   If there were collisions, it compares the provided key with the keys stored in that bucket (or follows the probing sequence) until it finds an exact match. Once the key is found, its associated value is returned.

### Why Dictionary Lookups Are Faster Compared to Lists or Tuples:

The difference in lookup speed comes down to how elements are organized and accessed in memory:

1.  **Lists and Tuples (Sequential Search)**:
    *   **Structure**: Both lists and tuples store elements in a defined order, typically in contiguous blocks of memory.
    *   **Lookup by Value**: If you need to find an item in a list or tuple by its *value* (e.g., `if 'apple' in my_list`), Python has to perform a **linear search**. This means it checks each element one by one from the beginning until it finds a match or reaches the end. In the worst case, it has to examine every element.
    *   **Time Complexity**: The average and worst-case time complexity for searching by value in lists and tuples is **O(n)**, where 'n' is the number of elements. This means lookup time grows proportionally with the size of the collection.
    *   **Lookup by Index**: Accessing an element by its *index* (e.g., `my_list[0]`) is very fast **O(1)** because Python can directly calculate its memory address.

2.  **Dictionaries (Hash-Based Direct Access)**:
    *   **Structure**: Dictionaries use hash tables, which are designed for direct access based on a key's hash value.
    *   **Lookup by Key**: When you look up a value by its key (e.g., `my_dict['key']`), Python calculates the hash value of the key and uses it to go almost directly to the memory location where the value is stored. It's like having a direct address book for each item.
    *   **Time Complexity**: On **average**, dictionary lookups by key have a time complexity of **O(1)**. This means the time it takes to find an item is roughly constant, regardless of how many items are in the dictionary.
    *   **Worst-Case**: In a worst-case scenario with many hash collisions, the lookup could degrade to O(n) (if Python has to search through a long linked list in a bucket). However, Python's hash table implementation is very sophisticated and engineered to minimize collisions, making average O(1) performance highly reliable.

**5. What is the difference between sets and lists in Python?
   How do sets handle duplicate values, and where are sets useful in practical scenarios?**

### Differences Between Sets and Lists in Python:

| Feature           | List (`list`)                                     | Set (`set`)                                            |
| :---------------- | :------------------------------------------------ | :----------------------------------------------------- |
| **Order**         | Ordered (elements maintain their insertion order) | Unordered (elements do not have a defined order)     |
| **Mutability**    | Mutable (can be changed after creation)           | Mutable (can be changed by adding/removing elements)   |
| **Duplicates**    | Allows duplicate elements                         | Does NOT allow duplicate elements (stores only unique items) |
| **Indexing**      | Supports indexing and slicing                     | Does NOT support indexing or slicing (unordered)     |
| **Element Type**  | Can store heterogeneous data types                | Can store heterogeneous data types, but elements must be hashable |
| **Syntax**        | Defined using square brackets `[]`                | Defined using curly braces `{}` or `set()` constructor |
| **Performance**   | Slower for membership testing (`in` operator)     | Faster for membership testing (average O(1) complexity) |

### How Sets Handle Duplicate Values:

The most fundamental characteristic of a set is that it **only stores unique elements**. When you try to add a duplicate value to a set, it simply ignores the addition; the set remains unchanged, and no error is raised.

**Example:**

```python
my_list = [1, 2, 2, 3, 4, 4, 5]
print(f"List with duplicates: {my_list}")

my_set = set(my_list) # Convert list to set
print(f"Set with unique values: {my_set}")

my_set.add(3) # Trying to add a duplicate
print(f"Set after adding a duplicate (3): {my_set}")

my_set.add(6) # Adding a new unique value
print(f"Set after adding a unique value (6): {my_set}")
```

### Where Sets Are Useful in Practical Scenarios:

Sets are powerful and efficient for tasks involving uniqueness and mathematical set operations. Here are some practical applications:

1.  **Removing Duplicates**: This is the most common use case. If you have a list or any collection with potential duplicate items and you need only the unique ones, converting it to a set and back to a list (if order is not important) is a very efficient way.
    *   **Example**: Getting all unique user IDs from a log file, or unique words from a document.

    ```python
    data = [10, 20, 30, 20, 10, 40, 50, 30]
    unique_data = list(set(data))
    print(f"Original data: {data}")
    print(f"Unique data: {unique_data}")
    ```

2.  **Membership Testing (Checking for Presence)**: Due to their hash-table-based implementation (similar to dictionary keys), checking if an element exists in a set is extremely fast, with an average time complexity of O(1). This is significantly faster than checking for membership in a list (which is O(n)).
    *   **Example**: Quickly checking if a username is already taken, or if a particular item is in a large inventory.

    ```python
    large_set = set(range(1_000_000))
    large_list = list(range(1_000_000))

    # Checking membership in a set is much faster
    import time

    start_time = time.time()
    123456 in large_set
    end_time = time.time()
    print(f"Time for set membership: {end_time - start_time:.6f} seconds")

    start_time = time.time()
    123456 in large_list
    end_time = time.time()
    print(f"Time for list membership: {end_time - start_time:.6f} seconds")
    ```

3.  **Mathematical Set Operations**: Sets provide efficient methods for standard set operations:
    *   `union()` or `|`: All unique elements from both sets.
    *   `intersection()` or `&`: Elements common to both sets.
    *   `difference()` or `-`: Elements in the first set but not in the second.
    *   `symmetric_difference()` or `^`: Elements in either set, but not in both.

    *   **Example**: Finding common interests between two users, identifying elements present in one group but not another, or checking if two groups are disjoint.

    ```python
    team_a = {'Alice', 'Bob', 'Charlie', 'David'}
    team_b = {'Charlie', 'Eve', 'Frank', 'Bob'}

    print(f"Team A: {team_a}")
    print(f"Team B: {team_b}")
    print(f"All team members (Union): {team_a.union(team_b)}")
    print(f"Common members (Intersection): {team_a.intersection(team_b)}")
    print(f"Members in Team A but not B (Difference): {team_a.difference(team_b)}")
    print(f"Members unique to each team (Symmetric Difference): {team_a.symmetric_difference(team_b)}")
    ```

4.  **Checking Subsets and Supersets**: Efficiently determine if one set is a subset or superset of another (`issubset()`, `issuperset()`).

**In summary**: Use lists when order matters, duplicates are allowed, and you need to access elements by index. Use sets when you need to store unique items, the order doesn't matter, and you perform frequent membership tests or mathematical set operations.

### 6. Write a program to count vowels in a given string.

````
def count_vowels(input_string):
    vowels = "aeiouAEIOU"
    count = 0
    for char in input_string:
        if char in vowels:
            count += 1
    return count

my_string_vowels = "Hello World"
vowel_count = count_vowels(my_string_vowels)
print(f"The string '{my_string_vowels}' has {vowel_count} vowels.")
````

### 7. Perform insertion and deletion operations on a list of integers.

````
numbers = [10, 20, 30, 40, 50]
print(f"Original list: {numbers}")

# Insertion: Add a new element at a specific index
numbers.insert(2, 25) # Inserts 25 at index 2
print(f"After insertion: {numbers}")

# Deletion: Remove an element by value
numbers.remove(40) # Removes the first occurrence of 40
print(f"After deletion by value: {numbers}")

# Deletion: Remove an element by index using pop()
popped_value = numbers.pop(0) # Removes and returns the element at index 0
print(f"After deletion by index (popped {popped_value}): {numbers}")
````

### 8. Convert a list into a tuple and a tuple into a list.

````
my_original_list = [1, 2, 3]
my_original_tuple = ('a', 'b', 'c')

# Convert list to tuple
list_to_tuple = tuple(my_original_list)
print(f"Original list: {my_original_list}, Converted to tuple: {list_to_tuple}")

# Convert tuple to list
tuple_to_list = list(my_original_tuple)
print(f"Original tuple: {my_original_tuple}, Converted to list: {tuple_to_list}")
````
