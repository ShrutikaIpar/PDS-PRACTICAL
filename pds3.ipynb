2.**You receive messy data with missing values, duplicates, and wrong types. What steps will you follow to clean it?**

**ANS** : When faced with messy data containing missing values, duplicates, and wrong data types, here are the general steps to follow for cleaning:

**Step 1: Understand the Data and Identify Issues**
*   **Load the Data:** Load the data into a suitable data structure (e.g., Pandas DataFrame in Python).
*   **Initial Inspection:** Use functions like `.info()`, `.describe()`, `.head()`, `.tail()` to get an overview of the data, column types, and basic statistics.
*   **Visualize Data:** Use plots (histograms, box plots) to identify outliers, skewed distributions, and potential errors.

**Step 2: Handle Missing Values**
*   **Identify Missing Values:** Use methods like `.isnull().sum()` to count missing values per column.
*   **Understand the Cause:** Determine if missing values are random, systematic, or due to data entry errors.
*   **Strategies for Handling Missing Values:**
    *   **Imputation:** Fill missing values with a statistical measure (mean, median, mode) for numerical data, or a common category for categorical data. Be cautious, as imputation can introduce bias.
    *   **Forward/Backward Fill:** For time-series or sequential data, fill missing values with the previous or next valid observation.
    *   **Deletion:** Remove rows or columns with a high percentage of missing values, but only if it doesn't lead to significant data loss or bias.
    *   **Advanced Techniques:** Use machine learning models (e.g., K-Nearest Neighbors imputation) for more sophisticated imputation.

**Step 3: Correct Data Types**
*   **Identify Incorrect Types:** Check column data types using `.info()`.
*   **Convert Types:** Convert columns to their correct data types (e.g., convert strings to numeric, objects to datetime, or numeric to categorical).
    *   Use `pd.to_numeric()`, `pd.to_datetime()`, `astype()`.
    *   Handle errors during conversion (e.g., `errors='coerce'` to turn unparseable values into NaN).

**Step 4: Remove Duplicate Data**
*   **Identify Duplicates:** Use `df.duplicated().sum()` to find duplicate rows.
*   **Decide on Duplicates:** Determine if duplicates are intentional or errors. Sometimes, partial duplicates (duplicates based on a subset of columns) need to be considered).
*   **Remove Duplicates:** Use `df.drop_duplicates()` to remove duplicate rows, keeping either the first or last occurrence.

**Step 5: Address Inconsistent Data and Formatting Issues**
*   **Text Data Cleaning:**
    *   **Standardize Case:** Convert text to lowercase or uppercase.
    *   **Remove Whitespace:** Trim leading/trailing spaces.
    *   **Correct Typos:** Use string methods (`.replace()`, regular expressions) or fuzzy matching for common misspellings.
    *   **Standardize Categories:** Ensure consistent naming for categorical variables (e.g., 'USA' vs 'U.S.A.' vs 'United States').
*   **Numerical Data Cleaning:**
    *   **Handle Outliers:** Identify and decide how to treat outliers (remove, cap, transform).
    *   **Standardize Units:** Ensure all numerical values in a column use the same unit (e.g., all in meters, not some in meters and some in centimeters).
*   **Date Data Cleaning:**
    *   **Standardize Formats:** Ensure all dates are in a consistent format.
    *   **Handle Invalid Dates:** Correct or remove dates that are out of a valid range.

**Step 6: Data Validation and Verification**
*   **Apply Business Rules:** Check if data conforms to known business rules or constraints.
*   **Cross-Reference:** Compare cleaned data with other reliable sources if available.
*   **Summarize Changes:** Keep track of the changes made during the cleaning process.

**Step 7: Document Cleaning Steps**
*   Maintain a clear record of all data cleaning steps taken. This ensures reproducibility and understanding for future analysis or team members.

5. **How do you reduce memory usage in pandas DataFrames?**

**ANS** : Reducing memory usage in pandas DataFrames is crucial for working with large datasets efficiently. Here are several effective strategies:

**1. Choose Appropriate Data Types (Downcasting):**
   *   Pandas often defaults to `int64` or `float64` for numerical columns and `object` for strings, which can be overkill for the actual range of values or cardinality. You can 'downcast' these to smaller types.
   *   **Integers:** If a column contains only integers between -128 and 127, it can be `int8` instead of `int64`.
   *   **Floats:** If precision is not paramount, `float32` can often replace `float64`.
   *   **Booleans:** Use `bool` type for true/false values.
   *   **Categorical Data:** For columns with a limited number of unique string values (e.g., 'Gender', 'Country'), converting them to the `category` data type can significantly reduce memory, as pandas stores an integer representation and a mapping of categories.

   ```python
   # Example of downcasting numerical types
   for col in df.select_dtypes(include=['int64']).columns:
       df[col] = pd.to_numeric(df[col], downcast='integer')
   for col in df.select_dtypes(include=['float64']).columns:
       df[col] = pd.to_numeric(df[col], downcast='float')

   # Example of converting to 'category' type
   for col in df.select_dtypes(include=['object']).columns:
       num_unique_values = len(df[col].unique())
       num_total_values = len(df[col])
       if num_unique_values / num_total_values < 0.5: # A heuristic: if unique values are less than 50% of total values
           df[col] = df[col].astype('category')
   ```

   *   As mentioned, converting `object` (string) columns with low cardinality to `category` is very effective.
   *   For string columns with high cardinality (many unique strings), if they are purely textual and not used for direct filtering or grouping, they will remain as `object` dtype, which consumes more memory. ### 11. Perform pivot and melt operations on a dataset.

````python
# Pivot: Reshape data based on column values
df_pivot = df.pivot_table(index='D', columns='B', values='A', aggfunc='sum')
print("Pivot Table (sum of 'A' by 'D' and 'B'):")
display(df_pivot)

# Melt: Unpivot a DataFrame from wide to long format
df_melt = df.melt(id_vars=['A', 'B'], value_vars=['C', 'F'], var_name='Metric', value_name='Value')
print("Melted DataFrame (unpivoting 'C' and 'F'):")
display(df_melt.head())
````

### 13. Compare merge v/s join operations.

````python
df_left = pd.DataFrame({'key': ['A', 'B', 'C', 'D'], 'value_left': np.random.randint(1, 10, 4)})
df_right = pd.DataFrame({'key': ['B', 'D', 'E', 'F'], 'value_right': np.random.randint(10, 20, 4)})

print("DataFrame Left:")
display(df_left)
print("DataFrame Right:")
display(df_right)

# Merge (explicit columns)
merged_df_ex = pd.merge(df_left, df_right, on='key', how='inner')
print("\nMerged DataFrame (inner merge on 'key' column):")
display(merged_df_ex)

# Join (index-based)
df_left_indexed = df_left.set_index('key')
df_right_indexed = df_right.set_index('key')
joined_df = df_left_indexed.join(df_right_indexed, how='inner')
print("\nJoined DataFrame (inner join on index):")
display(joined_df)

print("\nExplanation: `merge()` is a function that can merge DataFrames based on columns or indices (explicitly specified) and is more flexible for various types of joins (one-to-one, many-to-one, many-to-many). `join()` is a method primarily used for merging DataFrames based on their indices, though it can take an 'on' parameter for column-based joins. For complex joining criteria, `merge()` is generally preferred.")
````

### 14. Create custom aggregation functions.

````python
def range_agg(series):
    return series.max() - series.min()

custom_agg_df = df.groupby('B')['C'].agg(['min', 'max', 'mean', range_agg])
print("DataFrame with custom aggregation (range_agg on 'C' by 'B'):")
display(custom_agg_df)
````

### 15. Perform time series based data wrangling.

````python
# Set 'E' as the DataFrame index and sort it for time-series operations
df_ts = df.set_index('E').sort_index()
print("Original DataFrame with time index:")
display(df_ts)

# Resample daily and sum 'F'
df_resampled_daily = df_ts.resample('D')['F'].sum().reset_index()
print("\nResampled Daily Sum of 'F':")
display(df_resampled_daily)

# Calculate a 2-day rolling mean for 'F'
df_ts['F_rolling_mean'] = df_ts['F'].rolling('2D').mean()
print("\nDataFrame with 2-day rolling mean for 'F':")
display(df_ts)
````
