**AIM** : Implementing Web Scraping in Python with BeautifulSoup.

**1. What is web scraping? Explain the role of the Requests library in web scraping.**

**ANS** : Web scraping is the automated process of extracting data from websites. It involves programmatically fetching web pages and parsing their content to gather specific information.

The Requests library in Python plays a crucial role in web scraping by handling the process of making HTTP requests to retrieve web pages. It simplifies sending various types of HTTP requests (GET, POST, PUT, DELETE, etc.) and allows easy access to the server's response, including the HTML content of the page, headers, and status codes. In essence, Requests is the tool that fetches the raw web page content before it can be parsed and scraped.

**2. What is BeautifulSoup? How does it help in parsing HTML?**

**ANS **: BeautifulSoup is a Python library that is widely used for web scraping purposes to pull data out of HTML and XML files. It creates a parse tree for parsed pages that can be used to extract data from HTML, which is useful for web scraping.

BeautifulSoup helps in parsing HTML by:

Providing a parse tree: It takes the raw HTML content (often obtained using the Requests library) and turns it into a tree-like structure of Python objects. This structure represents the nested hierarchy of HTML tags, making it easy to navigate and search.

Offering intuitive methods for navigation and searching: It provides various methods to navigate the parse tree, such as by tag name, attributes, CSS selectors, or text content. This allows you to precisely locate and extract the specific pieces of data you need from a web page, even if the HTML is not perfectly formatted.

Handling malformed HTML: It's robust and can often work with imperfect HTML, which is common on the web, gracefully handling missing tags or other structural issues.

**3. Differentiate between web scraping and web crawling.**

**ANS** :  Web Crawling is primarily focused on discovering and indexing web pages. A web crawler (or spider) systematically browses the internet, following hyperlinks from page to page to create a 'map' of the web or a specific website. Its main function is navigation and discovery, building a list of URLs and understanding the site's structure, often for search engine purposes.


Web Scraping, on the other hand, is about extracting specific, structured data from web pages. Once a page is identified (sometimes by a crawler), a web scraper parses its HTML or XML content to pull out predefined pieces of information, such as product prices, news articles, or contact details. Its core function is data extraction, transforming unstructured web content into usable, structured data.

**4. What are the commonly used methods in BeautifulSoup?**

**ANS** :
 some commonly used methods in BeautifulSoup are:

 find(): Finds the first matching tag.

find_all(): Finds all matching tags.


select(): Finds tags using CSS selectors.

select_one(): Finds the first tag using CSS selectors.

.get_text(): Extracts all text content from a tag.

.name: Returns the tag's name.

.attrs: Returns a dictionary of the tag's attributes.

['attribute_name']: Accesses a specific attribute's value.

.contents, .children,

.descendants: For navigating down the parse tree.

.parent, .parents: For navigating up the parse tree.

**5.	What are the ethical and legal considerations in web scraping?**

**ANS** : ethical considerations for web scraping include respecting robots.txt, minimizing server load, considering data usage, protecting privacy, and being transparent. Legal considerations involve abiding by copyright laws, website Terms of Service, data protection regulations (like GDPR and CCPA), avoiding digital trespass, and preventing unfair competition.





```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time

# Base URL for the website
base_url = 'http://quotes.toscrape.com'
quotes_data = [] # List to store all scraped quotes

# Loop through multiple pages (e.g., first 10 pages)
for page_num in range(1, 11): # Adjust range as needed
    url = f'{base_url}/page/{page_num}/'
    print(f'Scraping page: {url}')

    response = requests.get(url)

    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')

        # Find all quote containers on the page
        quotes = soup.find_all('div', class_='quote')

        if not quotes: # No more quotes found, break the loop
            print(f'No quotes found on page {page_num}. Ending scraping.')
            break

        for quote in quotes:
            text = quote.find('span', class_='text').get_text()
            author = quote.find('small', class_='author').get_text()

            # Extract tags (optional)
            tags_elements = quote.find('div', class_='tags').find_all('a', class_='tag')
            tags = [tag.get_text() for tag in tags_elements]

            quotes_data.append({
                'Text': text,
                'Author': author,
                'Tags': ', '.join(tags)
            })

        # Optional: Add a small delay to be polite to the server
        time.sleep(1)
    else:
        print(f'Failed to retrieve page {url}. Status code: {response.status_code}')
        break

# Convert the list of dictionaries to a pandas DataFrame
df_quotes = pd.DataFrame(quotes_data)

# Display the first few rows of the DataFrame
display(df_quotes.head())

# Display some basic info about the collected data
print(f'\nTotal quotes scraped: {len(df_quotes)}')
print('\nDataFrame Info:')
df_quotes.info()
```

This program:

1.  **Imports necessary libraries**: `requests` for making HTTP requests, `BeautifulSoup` for parsing HTML, `pandas` for data manipulation, and `time` for introducing delays.
2.  **Initializes an empty list** (`quotes_data`) to store the extracted information.
3.  **Loops through multiple pages**: It constructs URLs for pages 1 to 10 (you can adjust the range) of the `quotes.toscrape.com` website.
4.  **Fetches and Parses**: For each page, it sends a GET request and parses the HTML content.
5.  **Extracts Data**: It identifies quote elements (`div` with class `quote`) and extracts the `Text`, `Author`, and `Tags` for each quote.
6.  **Stores Data**: Each extracted quote is stored as a dictionary in the `quotes_data` list.
7.  **Polite Scraping**: A `time.sleep(1)` is included to pause for 1 second between page requests, which helps prevent overwhelming the server and reduces the risk of being blocked.
8.  **Handles Pagination**: The loop breaks if no more quotes are found on a page.
9.  **Creates DataFrame**: After scraping all desired pages, the collected data is converted into a `pandas.DataFrame` for easy analysis and storage.
10. **Displays Results**: The first 5 rows of the DataFrame and its info are displayed.
